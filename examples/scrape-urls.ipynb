```python
websites = [
    "https://docs.wized.com/",
    "https://v1.wized.com/"
]
```

```python
from rag_scraper.link_extractor import LinkExtractor

page_urls = []
for url in websites:
    scraped_urls = LinkExtractor.scrape_url(url)
    
    page_urls.extend(scraped_urls)
```
